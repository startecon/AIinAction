```python
# define query pipeline
pipeline = [
  {
    '$vectorSearch': {
      'index': 'wind_power_vector_index', 
      'path': 'vectors', 
      'queryVector': vector, 
      'numCandidates': 100, 
      'limit': 10
    }
  }, {
    '$project': {
      '_id': 0, 
      'id': 1, 
      'date': 1, 
      'windpower': 1, 
      'score': {
        '$meta': 'vectorSearchScore'
      }
    }
  }
]
```

```python
# run the pipeline
result = coll.aggregate(pipeline)
```

```python
def df_to_vector(df, d, t, tp):
    df2 = df[
                (df["start time 3h interval (utc)"] == tp[t]) & 
                (df["day"] == d)
            ]
    arr = []
    arr.extend(df2["u-component wind in m/s"].to_list())
    arr.extend(df2["v-component wind in m/s"].to_list())
    arr.extend(df2["temperature in °C"].to_list())
    arr.extend(df2["global surface radiation in W/m2"].to_list())

    return arr
```

```python
def get_history_point(h, d, t):
    if (t-h > -1):
        return d, t
    else:
        f = math.floor((h-1)/8)+1
        return d-f, f*8+(t-h)
```
